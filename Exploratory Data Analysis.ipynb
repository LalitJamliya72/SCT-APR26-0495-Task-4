```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("C:\\Users\\HP\\Downloads\\Marketing_Campaign_Data.xlsx")
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
# Missing values
df.isnull().sum()

# Drop missing
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()
```

```python
df['ROI'] = (df['Revenue (₹)'] - df['Cost (₹)']) / df['Cost (₹)']
```

```python
df['Conversion_Rate'] = df['Impressions'] / df['Clicks']
```

```python
df.groupby('Platform')['Revenue (₹)'].sum().plot(kind='bar')
plt.title("Revenue by Channel")
plt.show()
```

```python
df.groupby('Platform')['ROI'].mean().plot(kind='bar')
plt.title("ROI by Channel")
plt.show()
```

```python
funnel = df[['Impressions','Clicks','Leads']].sum()
funnel.plot(kind='bar')
plt.title("Marketing Funnel")
plt.show()
```
